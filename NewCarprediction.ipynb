#cleaning the data

```python
car =car[car['year'].str.isnumeric()]
```

```python
print(car) # year values are filtered
```

```python
car['year']=car['year'].astype(int) #converted the year object to int
```

```python
car.info() # year data is cleaned
```

```python
# cleaning price column
car = car[car['Price']!='Ask For Price'] # removing string value
```

```python
car['Price']= car['Price'].str.replace(',','').astype(int)
```

```python
car.info()
```

```python
#cleaning kms_driven column
car['kms_driven']
```

```python
car['kms_driven']=car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car =car[car['kms_driven'].str.isnumeric()]
print(car['kms_driven'])
```

```python
car['kms_driven']=car['kms_driven'].astype(int)
car.info()
```

```python
car = car[~car['fuel_type'].isna()]
```

```python
#cleaning the name column
car['name']= car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car
```

```python
#some row are drop by this so we have to reset the index in order
car = car.reset_index(drop=True)
```

```python
car = car[car['Price']<6e6].reset_index(drop=True)
```

#Trannig the model

```python
X = car.drop(columns='Price')
Y = car['Price']
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder() # this will encode the data categorical data into numeric data
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
linr_model = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,linr_model)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_predict=pipe.predict(X_test)
```

```python
print(y_predict)
```

```python
scores =[]
for i in range(1000):
  X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=i)
  lr = LinearRegression()
  pipe = make_pipeline(column_trans,lr)
  pipe.fit(X_train,Y_train)
  y_predict = pipe.predict(X_test)
  scores.append(r2_score(Y_test,y_predict))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,Y_train)
y_predict = pipe.predict(X_test)
r2_score(Y_test,y_predict)
```

```python
import pickle
pickle.dump(pipe,open('LinearRegressionPrediction.pkl','wb'))
```
